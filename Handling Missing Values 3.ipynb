# 2. Filling Missing Values
- filna()
    - 1. mean()
    - 2. median()
    - 3. mode()
    - 4. min()
    - 5. max()
    - 6. Static Value
    

```python
df.info()
```

```python
df["Age in Yrs"].mean()
```

```python
df["Age in Yrs"].median()
```

```python
filled_df = df.fillna(100)
filled_df
```
