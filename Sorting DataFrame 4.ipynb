```python
demo_df = pd.DataFrame(
    {
        "Numbers" : np.random.randint(1,10,10),
        "Alpha" : np.random.choice(list("ABCD"),10)
     }
)
```

```python
demo_df
```

```python
demo_df.sort_values(["Numbers","Alpha"])
```

```python
demo_df.sort_values(["Alpha","Numbers"])
```

```python
demo_df.sort_values("Alpha",ascending=False)
```

```python
demo_df.sort_values(["Alpha","Numbers"],ascending=[True,False])
```

```python
demo_df
```

```python
demo_df.sort_values(["Alpha","Numbers"],ascending=[True,False],inplace=True)
```

```python
demo_df
```
