```python
import pandas as pd
import numpy as np
```

```python
movie_df = pd.read_csv("movies_data.csv")
movie_df.head()
```

```python
movie_df.groupby("genre")
```

```python
movie_df.groupby("genre").first()
```

```python
movie_df.groupby("genre").get_group("Action")
```

```python
action_df = movie_df.groupby("genre").get_group("Action")
action_df.reset_index(drop=True)
```

```python
drama_df = movie_df.groupby("genre").get_group("Drama")
drama_df.reset_index(drop=True)
```

```python
crime_df = movie_df.groupby("genre").get_group("Crime")
crime_df.reset_index(drop=True)
```
