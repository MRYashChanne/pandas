# Create Dictionary

```python
df = pd.DataFrame()
df["Numbers"] = [1,2,3,4,5]
```

```python
df
```

```python
df["Square"] = df["Numbers"] ** 2
df["Cube"] = df["Numbers"] ** 3
df
```

```python
df.to_dict()
```

```python
df.to_dict(orient="list")
```

```python
df.to_dict(orient="records")
```
