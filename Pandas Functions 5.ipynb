```python
Principle_Amount(P) = [100000,200000,500000,700000,10000000]
Interest_Rate(R) = [11,12,14,13,7]
Tenure(T) = [2,4,5,3,7]

Amount = P(1 + R / 100) ^ T
Coumpound_Interest = Amount - Principle_Amount
```

```python
df = pd.DataFrame()
df["P"] = [100000,200000,500000,700000,10000000]
df["R"] = [11,12,14,13,7]
df["T"] = [2,4,5,3,7]
df["Amount"] = df["P"]*((1+(df["R"]/100))**df["T"])
df["Compound_interest"] = df["Amount"] - df["P"]
df
```

```python
dict1 = {
    "Principle_Amount(P)": [100000,200000,500000,700000,10000000],
    "Interest_Rate(R)" : [11,12,14,13,7],
    "Tenure(T)" : [2,4,5,3,7]
}
df= pd.DataFrame(dict1)
df["Amount"]=df["Principle_Amount(P)"]*(1 + df["Interest_Rate(R)"] /100)**df["Tenure(T)"]
df["Coumpound_Interest"]= df["Amount"]- df["Principle_Amount(P)"]
df
```
