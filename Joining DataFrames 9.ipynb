# 3. merge()
- pd.merge(left_df,right_df,how)

```python
df1 = pd.DataFrame(
    {
        "Emp Name" : ["Suraj","Sagar","Pravin","Rohit"],
        "Age" : [30,29,25,35],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"],
        "Salary" : [30000,50000,60000,40000]
    },
    index = list("abcd")
)

df2 = pd.DataFrame(
    {
        "Emp Name" : ["Akshay","Suresh","Pravin","Rohit"],
        "Age" : [25,27,25,35],
        "Location" : ["Pune","Mumbai","Delhi","Chennai"],
        "Designation" : ["PD","DS","JD","SE"]
    },
    index = list("cdxy")
)
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how="outer")
```

```python
pd.merge(df1,df2,how="inner")
```

```python
pd.merge(df1,df2,how="left")
```

```python
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1,df2,on="Emp Name",how="outer",indicator=True)
```

```python
pd.merge(df1,df2,how="outer",indicator=True)
```

```python
pd.merge(df1,df2,on="Emp Name",how="inner",indicator=True)
```

```python
pd.merge(df1,df2,on="Emp Name",indicator=True)
```
